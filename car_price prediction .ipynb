```python
df_obj = df.select_dtypes(include = ['object']).copy()
df_obj.head()
```

```python
df_obj['CarName'].unique()
```

```python
car = df_obj.CarName.str.split(expand = True) 
Brand = car[0] 
df_obj['CarBrand'] = Brand 
df_obj.drop(columns = ['CarName'],inplace = True) 
df_obj.head()
```

```python
print(df_obj['doornumber'].value_counts()) 
print('\n')
print(df_obj['cylindernumber'].value_counts()) 
```

```python
encodedoorandcy = {"doornumber":     {"four": 4, "two": 2},
                "cylindernumber": {"four": 4, "six": 6, "five": 5, "eight": 8,
                                  "two": 2, "three": 3, "twelve":12 }}
df_obj = df_obj.replace(encodedoorandcy) 
df_obj.head()
```

```python
one_hot = pd.get_dummies(data = df_obj, 
                         columns=['carbody','drivewheel','fueltype',
                                  'aspiration','enginelocation'
                                  ,'fuelsystem','CarBrand'])
df_obj= one_hot 
df_obj.head()
```

```python
df_obj['enginetype'].value_counts()
```

```python
df_obj["enginetype"] = np.where(df_obj["enginetype"].str.contains("ohc"), 1, 0) 
df_obj.head()
```

```python
df_obj.dtypes
```

```python
nonobj_df =df.select_dtypes(include=['int64', 'float64']).copy()  
df=df_obj.join(nonobj_df)
```

```python
for i in df.columns:
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)
    IQR = Q3 - Q1
    LowerBound = Q1 - 1.5 * IQR
    UpperBound = Q3 + 1.5 * IQR
    df = df[(df[i] >= LowerBound)|(df[i] <= UpperBound)]
```

```python
X = df.drop(columns=['price', 'car_ID'])
y = df['price']


from sklearn.model_selection import train_test_split # import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

import pandas as pd 
import statsmodels.api as sm 

def backward_regression(X, y,
                           initial_list = [], 
                           threshold_in = 0.01, 
                           threshold_out = 0.05, 
                           verbose=True):
    included = list(X.columns)
    while True:
        changed=False
        model = sm.OLS(y, sm.add_constant(pd.DataFrame(X[included]))).fit()
        pvalues = model.pvalues.iloc[1:]
        worst_pval = pvalues.max() 
        if worst_pval > threshold_out:
            changed = True
            worst_feature = pvalues.idxmax()
            included.remove(worst_feature)
        if not changed:
            break
    return included

backward_regression(X_train, y_train)
```

```python
from sklearn.metrics import r2_score
yhat = regr.predict(X_test)

print("Mean absolute error: %.2f" % np.mean(np.absolute(yhat - y_test)))

print("Residual sum of squares (MSE): %.2f" % np.mean((yhat - y_test) ** 2)) 

print("R2-score: %.2f" % r2_score(y_test,yhat ))
```
